```python
# 1. import needed libraries, set plots to display in notebook

import numpy as np
from pandas import Series, DataFrame
%matplotlib inline
import matplotlib.pyplot as plots
# allows currency formatting
import locale
locale.setlocale(locale.LC_ALL, 'en_IN.UTF-8')
```

```python
# 2. A traditional savings calculator approach

pv = 100000
time_horizon = 30
i =0.13
additions = 100000

for year in range(time_horizon):
    ending = pv * (1+i) + additions
    print(locale.currency(ending, grouping=True))
    pv = ending
```

```python
# 3. Generating one possible future value based on market history; I will use 13% expected return with 18% volatility

pv = 100000
expected_return = .13
volatility = .18
time_horizon = 30
annual_addition = 100000

print("\tReturn", "\t\tEnding Value".rjust(18))
for year in range(time_horizon):
    market_return = np.random.normal(expected_return, volatility)
    fv = pv * (1 + market_return) + annual_addition
    print("\t{}".ljust(10).format(round(market_return,4)), "\t{}".rjust(10).format(locale.currency(fv, grouping=True)))
    pv = fv
```

```python
# 4. Simulate portfolio ending market values 

sim = DataFrame()
iterations = 5000


for x in range(iterations):
    expected_return = .13
    volatility = .18
    time_horizon = 30
    pv = 100000
    annual_investment = 100000
    stream = []
    for i in range(time_horizon):
        end = round(pv * (1 + np.random.normal(expected_return,volatility)) + annual_investment,2)
        
        stream.append(end)

        pv = end

    
    sim[x] = stream
```

```python
# 5. Sample first five streams of annual ending values
first_five = list(range(5))
sim[first_five]
```

```python
# 6. Plot first five simulated portfolios
plots.plot(sim[first_five])
plots.xlabel("Years")
plots.ylabel("Returns (in crores)")
```

# PRICING OPTIONS

```python
S=100 #initial price of underlying stock
T=1 #Time period for maturity of stock
r=0.07 # discount rate (from FD or gov bonds)
sigma=0.2 #volatility of underlying asset 20%
Nsimulations=5000
Nsteps =250 #250 trading days
dt=T/Nsteps
K=100 #strike price
```

```python
import numpy as np

drift=(r-(sigma**2)/2)*dt
a= sigma*np.sqrt(dt)
x=np.random.normal(0,1,(Nsimulations,Nsteps))
```

```python
Smat=np.zeros((Nsimulations,Nsteps))
Smat[:,0]+=S

for i in range(1,Nsteps):
    Smat[:,i]+= Smat[:,i-1]*np.exp(drift+a*x[:,i])
```

```python
q=Smat[:,-1]-K
for i in range(len(q)):
    if q[i]<0:
        q[i]=0
    else:
        q[i]=q[i]
```

```python
p=-Smat[:,-1]+K
for i in range(len(q)):
    if p[i]<0:
        p[i]=0
    else:
        p[i]=p[i]
```

```python
payoff_call=np.mean(q)
payoff_put=np.mean(p)
```
